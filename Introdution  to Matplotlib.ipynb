### Mataplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(5,10))

# 3. Plot data
ax.plot(x,y)

# 4. Customize plot
ax.set(title="Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("img/sample-plot.png")
```

## Marking figures with NumPy arrays

We want
* Line plot
* Scatter plot
* Bar plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# make a plot from dictionary
nut_butter_prices = {"Almond Butter": 10, "Peanut Butter": 8, "Cashew Butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), height=nut_butter_prices.values())
ax.set(title="Dan's Nut butter store", ylabel="Price ($)", xlabel="Butter Type")
```

```python
fig,ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()), )
```

```python
# make some data for histograms and plot it
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x)
```

### Two option for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax4.hist(np.random.randn(1000))
```

```python
# subplots option 2
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# Plot to each different index
ax[0, 0].plot(x, x/2)
ax[0, 1].scatter(np.random.random(10), np.random.random(10))
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax[1, 1].hist(np.random.randn(1000))
```

### Plotting from pandas DataFrame

```python
import pandas as pd
```

```python
car_sales = pd.read_csv("car-sales.csv")
car_sales
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range("1/1/2020", periods=1000))
ts = ts.cumsum()
ts.plot()
```

```python
car_sales
```

```python
# Ensure the 'Price' column is of string type
car_sales["Price"] = car_sales["Price"].astype(str)

# Remove dollar signs, commas, and periods
car_sales["Price"] = car_sales["Price"].str.replace('[\$\,\.]', '', regex=True)

# Display the DataFrame
car_sales["Price"] = car_sales["Price"].str[:-2].astype(float)
car_sales
```

```python
car_sales["Sale Date"] = pd.date_range("1/1/2023", periods=len(car_sales))
car_sales
```

```python
car_sales["Total Sales"] = car_sales["Price"].cumsum()
car_sales
```

```python
# Let's plot the total sales
car_sales.plot(x="Sale Date", y="Total Sales")
```

```python
car_sales.plot(x="Odometer (KM)", y="Price", kind="scatter")
```
